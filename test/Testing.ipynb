## Live 1 out testing

```python
import os
import pandas as pd
```

```python
def execute_command(command = ''): 
    stream = os.popen(command)
    return stream.read()

class CsvAction:
    def __init__(self, csv_path):
        self.csv = pd.read_csv(csv_path, sep=",", names=["to", "from"],
                                   header=0)
        self.list_of_unique_papers = self.csv["to"].unique()
        self.unique_number = len(self.list_of_unique_papers)

    def remove_one(self, i=1):
        new_csv = self.csv[self.csv['to'] != self.list_of_unique_papers[i]]
        new_csv.to_csv("tmp.txt", index=False, header=False)
        return "tmp.txt"
```

### Running algorithms
Please do not restart this section, it may take time and the data is already collected. For the analysis, please scroll down. 

```python
def record_results(inital_dataset_name):
    for method in ['network_degree', 'network_betweenness', 'jaccard', 'jaccard_no_division', 'paper']:
        ca = CsvAction(inital_dataset_name)

        print(ca.unique_number)
        for i in range(0, ca.unique_number):
            dataset_name = ca.remove_one(i=i)
            tmp = execute_command("python3 analysis.py -m "+method+" -i " +dataset_name+" -o TestResults/test_"+method+"_"+str(i)+"_"+inital_dataset_name+" -n 12")
            print(tmp)
            print("---------------")
```

```python
inital_dataset_name = 'output_paper1.csv'
record_results(initial_dataset_name)
```

### Analysis

```python
for method in ['network_degree', 'network_betweenness', 'jaccard', 'jaccard_no_division', 'paper']:
    result_pos = 0
    result_total = 0
    for inital_dataset_name in ['output_paper1.csv','output_paper2.csv','output_paper3.csv','output_paper4.csv','output_paper5.csv']:
        ca = CsvAction(inital_dataset_name)


        for i in range(0, ca.unique_number):
            result_name = "TestResults/test_"+method+"_"+str(i)+"_"+inital_dataset_name
            res_csv = pd.read_csv(result_name, sep=",", names=["res"],
                                       header=0)
            for paper in res_csv['res'].values:
                if ca.list_of_unique_papers[i] == paper:
                    result_pos = result_pos+1
                    break
            result_total = result_total+1

    print(method +": pos <{"+str(result_pos)+"}> total <{"+str(result_total)+"}>")
    print("\t Recall: <{"+ str(round(result_pos/ result_total, 3)) +"}>" )
```
